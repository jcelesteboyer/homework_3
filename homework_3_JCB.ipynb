# **The Unix Shell**

1. The first thing I learned in the unix shell readings is that following the ls command with a -F will tell you more information about what are the types of files listed. The -F is known as a 'switch' or 'flag'. The full command will tell you all the files like normal but have an addition at the end such as '*', '/', or '@'. The '/' means it is a directory, the '*' means it is executable, and the '@' means that it is a link. If the file is missing a symbol, then it is a normal file like a .txt. 
2. The 'mv' command can help you move files from different directories. You need to put 'mv' in front of the file you want to move followed by the destination directory. The destination directory should start with either a '.' to mean current directory or '..' to mean the parent directory. Follow that with a '/' and the name of the destination directory. 
3. The '|' is called a pipeline. When you use that vertical bar symbol you are saying that you want the output of the command on the left side to be the input of the command on the right side.
